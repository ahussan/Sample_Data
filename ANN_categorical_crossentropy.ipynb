Artificial Neural Network(ANN) using keras framework

```python
#Dependencies
import numpy as np
import pandas as pd
#dataset import
train_df = pd.read_csv("/TrainResult.csv", skiprows = 0)

test_df = pd.read_csv("/TestResult.csv", skiprows = 0)
```

```python
#Changing pandas dataframe to numpy array
Xtrain = train_df.iloc[:,:15].values
Ytrain = train_df.iloc[:,15:16].values

Xtest = test_df.iloc[:,:15].values
Ytest = test_df.iloc[:,15:16].values
```

```python
import tensorflow
Ytrain = tensorflow.keras.utils.to_categorical(Ytrain, 2)
Ytest = tensorflow.keras.utils.to_categorical(Ytest, 2)
```

```python
#Dependencies
import keras
from keras.models import Sequential
from keras.layers import Dense
```

# Model loss function: categorical_crossentropy

##### **optimizer='adam'**

```python
# Neural network
model = Sequential()
model.add(Dense(12, input_dim=15, activation='relu')) #1st hidden Layer
model.add(Dense(8, activation='relu')) #2nd hidden Layer
model.add(Dense(2, activation='softmax')) #Final Layer
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(Xtrain, Ytrain, epochs=200, batch_size=40)
```

```python
#Testing
y_pred = model.predict(Xtest)
#Converting predictions to label
pred = list()

for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
    
#Converting one hot encoded test label to label
test = list()
for i in range(len(Ytest)):
    test.append(np.argmax(Ytest[i]))
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
a = accuracy_score(pred,test)
print('Accuracy of ANN Multiclass Classification with optimizer=adam:', a*100)
```
